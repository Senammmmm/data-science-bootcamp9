# **HOMEWORK 01 - Star Wars API**

- write API to extract data
- create data frame and write it to CSV file

```python
# import module
from requests import get
from time import sleep
import pandas as pd
```

```python
url ="https://swapi.py4e.com/api/people/"

## Empty list to store data
data = []

## homework API 01
for i in range(5):
    index = i + 1
    new_url = url + str(index)
    resp = get(new_url).json()
    name = resp["name"]
    height = resp["height"]
    weight = resp["mass"]
    gender = resp.get("gender")

    ## Handle missing gender data
    if gender == "n/a":
        gender = "would rather not say"
    else:
        gender = gender

    homeworld = resp["homeworld"]


    ## Append data to the list as a dictionary
    data.append({
        "name": name,
        "height": height,
        "weight": weight,
        "gender": gender,
        "homeworld": homeworld
    })
    sleep(2)

## Create DataFrame from the data list
df1 = pd.DataFrame(data)
df1
```

```python
url_pla ="https://swapi.py4e.com/api/planets/"

## Empty list to store data
data_homeworld = []

## homework API 01
for i in range(8):
    index = i + 1
    new_url_pla = url_pla + str(index)
    resp_pla = get(new_url_pla).json()
    homeworld = url_pla + str(index) + "/"
    homeworld_name = resp_pla["name"]

    ## Append data to the list as a dictionary
    data_homeworld.append({
        "homeworld": homeworld,
        "homeworld_name": homeworld_name
    })
    sleep(2)

## Create DataFrame
df2 = pd.DataFrame(data_homeworld)
df2
```

```python
## left join to find homeworld_name

df3 = df1.merge(df2, on="homeworld", how="left")
df3
```

```python
df4 = df3[['name', 'height', 'weight', 'gender', 'homeworld_name']]
df4
```

```python
## change dataframe column names

col_names = list(df4.columns)
change_names = [name.replace("homeworld_name","homeworld")
                    for name in col_names]
df4.columns = change_names
df4
```

```python
## write in CSV

df4.to_csv("StarWarsChar.CSV")
```

# **HOMEWORK 02 - Public API**

```python
from requests import get
import datetime

url_02 ="https://hp-api.onrender.com/api/characters/"

response = requests.get(url_02)

if response.status_code == 200:
    print("Ready!")
    data = response.json()
```

```python
## Empty list to store data
data_02 = []

# Initialize a counter
count = 0

# Current year
current_year = datetime.datetime.now().year

## homework API 01
for character in data:
    name = character["name"]
    gender = character["gender"]
    house = character["house"]
    year_of_birth = character["yearOfBirth"]
    age = current_year - character["yearOfBirth"]
    ancestry = character["ancestry"]
    species = character["species"]
    count += 1
    if count == 10: # Break the loop after 10 names
        break

    ## Append data to the list as a dictionary
    data_02.append({
        "name": name,
        "gender": gender,
        "house": house,
        "age": age,
        "ancestry": ancestry,
        "species": species
    })
    sleep(2)

# Create DataFrame from the data list
df5 = pd.DataFrame(data_02)
df5
```
